```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Read the data

```python
data = pd.read_csv('midterm.csv')
data = data.rename(columns={'SCF40_Temp (°C)':'SCD40_Temp (°C)'}) ## correcting tipo in Arduino code while naming SCD40
# pd.to_datetime(data['RTC'], format="%Y/%m%d %H:%M:%S")
data = data.drop([4], axis=0)
data['RTC'] = pd.to_datetime(data['RTC'], format="ISO8601") ## fixing date format & letting pandas know it's type date
data = data.rename(columns={'RTC':'Time'}) ## correcting tipo in Arduino code while naming SCD40
data = data.set_index('Time')
data = data.astype(float)
data
```

### Temperature

```python
p = data.iloc[:,[4,2,1]].plot(figsize=[8,5]).set_ylabel('Temperature °C')
fig = p.get_figure()
plt.legend()
plt.title('Comparison of temperature measured on 3 sensors')
fig.savefig('Output/Temp_Comparison.jpg');
```

### Relative Humidity

```python
p = data.iloc[:,[5,3]].plot(figsize=[8,5]).set_ylabel('RH %')
fig = p.get_figure()
plt.legend()
plt.title('Comparison of Relative Humidity measured on 2 sensors')
fig.savefig('Output/RH_Comparison.jpg');
```

### $CO{_2}$ Concentration

```python
p = data.iloc[:, 6].plot(figsize=[8,5]).set_ylabel('Concentration ppm')
fig = p.get_figure()
plt.legend()
plt.title('$CO{_2}$ Concentraion')
fig.savefig('Output/CO2.jpg');
```
